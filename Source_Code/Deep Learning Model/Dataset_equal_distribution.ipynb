```python
import pandas as pd

#Original 2015 dataset
df = pd.read_csv('Laptops_Train.csv')

target_size = 10000

classes = df['polarity'].unique().tolist()

base_count = target_size // len(classes)
remainder = target_size - base_count * len(classes)

new_counts = {cls: base_count for cls in classes}
for cls in classes[:remainder]:
    new_counts[cls] += 1

data = []
for polarity, count in new_counts.items():
    sampled_data = df[df['polarity'] == polarity].sample(n=count, replace=True)
    data.append(sampled_data)

new_df = pd.concat(data, ignore_index=True)
new_df = new_df.sample(frac=1).reset_index(drop=True)
new_df.to_csv('Laptops_Train_equal_distribution.csv', index=False)

print("Number of samples for each category:", new_df['polarity'].value_counts().to_dict())
```
